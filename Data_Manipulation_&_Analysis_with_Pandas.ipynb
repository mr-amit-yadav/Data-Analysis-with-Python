# Data Manipulation and Analysis with Pandas

Data Manipulation and Analysis are key tasks in any data science or data analysis project.<br>
Pandas provides a wide range of functions for Data Manipulation and Analysis,making it easier to clean,transform, and extract insights from data.<br>
In this lesson,we will cover various data manipulation and analysis techniques using Pandas.

```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
## Fetch the first 5 rows
df.head(5)
```

```python
df.tail(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
### Handling Missing Values
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_filled=df.fillna(0)
```

```python
### Filling missing values with the mean of the column
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming Columns
df=df.rename(columns={'Sale Date':'Sale Date'})
df.head()
```

```python
## Change datatypes
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['New Value']=df['Value'].apply(lambda x:x*2)
df.head(5)
```

```python
## Data Aggregating and Grouping
df.head(5)
```

```python
grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
df.groupby(['Product','Region'])['Value'].mean()
```

```python
## Aggregate multiple Functions
grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
### Merging and Joining DataFrames
# Create sample DataFrames
df1=pd.DataFrame({'key':['A','B','C'],'value':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge DatFrame on the 'Key Columns'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="right")
```

```python
pd.merge(df1,df2,on="key",how="left")
```
